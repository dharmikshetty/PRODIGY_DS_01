Create a bar chart or histogram to visualize the distribution of categorized or continous variabke such as the distribution of ages or genders in a population.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/sales.csv")
```

```python
df
```

#PIE CHART

```python
import matplotlib.pyplot as plt

# Grouping by Category and calculating the total sales for each category
category_sales = df.groupby('Category')['Total_Sale'].sum()

# Plotting a pie chart
plt.figure(figsize=(8, 8))
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Sales by Category')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

#HISTOGRAM

```python
# Plotting a histogram of unit prices
plt.figure(figsize=(8, 6))
plt.hist(df['Unit_Price'], bins=15, color='skyblue', edgecolor='black')
plt.xlabel('Unit Price')
plt.ylabel('Frequency')
plt.title('Histogram of Unit Prices')
plt.show()
```

#BOXPLOT

```python
import seaborn as sns

# Creating a box plot for quantity sold in each category
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Quantity_Sold', data=df)
plt.title('Quantity Sold by Category')
plt.show()
```
